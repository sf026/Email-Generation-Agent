```python
from crewai import Agent, Task, Crew, LLM
```

### BLOG

```python
llm = LLM(
    api_key=" ",
    model="groq/llama-3.1-8b-instant"
)
```

###  AGENTS

```python
email_agent = Agent(
    role="Email Assistant",
    goal="Write professional emails based strictly on the given purpose",
    backstory="You write short, clear, professional emails. Do not write anything unrelated.",
    llm=llm,
    allow_delegation=False,
    verbose=True
)
```

### TASKS

```python
email_task = Task(
    description=(
        "Write a professional email strictly for the following purpose:\n"
        "{purpose}\n\n"
        "Rules:\n"
        "- Use a proper subject line\n"
        "- Keep it concise\n"
        "- Do not change the intent\n"
        "- Do not invent unrelated content"
    ),
    expected_output="A professional email matching the given purpose.",
    agent=email_agent
)
```

### DEFINE CREW

```python
email_crew = Crew(
    agents=[email_agent],
    tasks=[email_task],
    verbose=True
)
```

### KickOff

```python
result = email_crew.kickoff(
    inputs={"purpose": "Requesting leave for two days due to personal reasons"})
```

### DISPLAY
